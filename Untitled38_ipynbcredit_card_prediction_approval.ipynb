```python
# ========================================
# CREDIT CARD APPROVAL PREDICTION PROJECT
# Single Cell Version for Google Colab
# ========================================

# STEP 1: Install Required Libraries
print("Installing required libraries...")
!pip install -q seaborn mlflow
print("Installation complete!\n")

# STEP 2: Import Libraries
print("Importing libraries...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import mlflow
import mlflow.sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
sns.set_style('whitegrid')
print("Libraries imported!\n")

# STEP 3: Load Dataset
print("Downloading UCI Credit Approval Dataset...")
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data"
filename = "/tmp/crx.data"
urllib.request.urlretrieve(url, filename)
column_names = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15', 'Target']
df = pd.read_csv(filename, names=column_names, na_values='?')
print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns\n")

# STEP 4: Quick Data Exploration
print("Dataset Overview:")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Target distribution:\n{df['Target'].value_counts()}\n")

# STEP 5: Data Preprocessing
print("Preprocessing data...")
# Separate features and target
X = df.drop('Target', axis=1)
y = df['Target']

# Identify column types
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Handle missing values and encode
X_processed = X.copy()
for col in categorical_cols:
    X_processed[col].fillna(X_processed[col].mode()[0], inplace=True)
    le = LabelEncoder()
    X_processed[col] = le.fit_transform(X_processed[col].astype(str))

for col in numerical_cols:
    X_processed[col].fillna(X_processed[col].mean(), inplace=True)

# Scale numerical features
scaler = StandardScaler()
X_processed[numerical_cols] = scaler.fit_transform(X_processed[numerical_cols])

# Encode target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
print(f"Preprocessing complete!")
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples\n")

# STEP 6: Train Models with MLflow Tracking
print("Training models...\n")
mlflow.set_experiment("credit-card-approval")

models = {}
accuracies = {}
predictions = {}

# Model 1: Logistic Regression
print("Training Logistic Regression...")
with mlflow.start_run(run_name="Logistic_Regression"):
    lr = LogisticRegression(max_iter=1000, random_state=42)
    lr.fit(X_train, y_train)
    y_pred_lr = lr.predict(X_test)
    acc_lr = accuracy_score(y_test, y_pred_lr)
    mlflow.log_params({"model_type": "Logistic Regression", "max_iter": 1000})
    mlflow.log_metric("accuracy", acc_lr)
    models['Logistic Regression'] = lr
    accuracies['Logistic Regression'] = acc_lr
    predictions['Logistic Regression'] = y_pred_lr
    print(f"✅ Accuracy: {acc_lr*100:.2f}%\n")

# Model 2: Decision Tree
print("Training Decision Tree...")
with mlflow.start_run(run_name="Decision_Tree"):
    dt = DecisionTreeClassifier(max_depth=5, random_state=42)
    dt.fit(X_train, y_train)
    y_pred_dt = dt.predict(X_test)
    acc_dt = accuracy_score(y_test, y_pred_dt)
    mlflow.log_params({"model_type": "Decision Tree", "max_depth": 5})
    mlflow.log_metric("accuracy", acc_dt)
    models['Decision Tree'] = dt
    accuracies['Decision Tree'] = acc_dt
    predictions['Decision Tree'] = y_pred_dt
    print(f"✅ Accuracy: {acc_dt*100:.2f}%\n")

# Model 3: Random Forest (Best Model)
print("Training Random Forest...")
with mlflow.start_run(run_name="Random_Forest"):
    rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
    rf.fit(X_train, y_train)
    y_pred_rf = rf.predict(X_test)
    acc_rf = accuracy_score(y_test, y_pred_rf)
    mlflow.log_params({"model_type": "Random Forest", "n_estimators": 100, "max_depth": 5})
    mlflow.log_metric("accuracy", acc_rf)
    models['Random Forest'] = rf
    accuracies['Random Forest'] = acc_rf
    predictions['Random Forest'] = y_pred_rf
    print(f"✅ Accuracy: {acc_rf*100:.2f}%\n")

# STEP 7: Model Comparison
print("="*60)
print("MODEL COMPARISON RESULTS")
print("="*60)
for model_name, acc in sorted(accuracies.items(), key=lambda x: x[1], reverse=True):
    print(f"{model_name:20s}: {acc*100:.2f}%")
best_model_name = max(accuracies, key=accuracies.get)
print(f"\n Best Model: {best_model_name} ({accuracies[best_model_name]*100:.2f}%)\n")

# STEP 8: Detailed Evaluation (Best Model)
print("="*60)
print(f"DETAILED EVALUATION - {best_model_name}")
print("="*60)
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, predictions[best_model_name])
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, predictions[best_model_name],
                          target_names=['Denied (-)', 'Approved (+)']))

# STEP 9: Feature Importance (for Random Forest)
if best_model_name == 'Random Forest':
    print("\n Top 5 Most Important Features:")
    feature_imp = pd.DataFrame({
        'Feature': X_processed.columns,
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)
    print(feature_imp.head(5).to_string(index=False))

# STEP 10: Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Model Comparison
model_names = list(accuracies.keys())
model_accs = [accuracies[m]*100 for m in model_names]
axes[0].bar(model_names, model_accs, color=['#FFD700', '#C0C0C0', '#CD7F32'])
axes[0].set_ylabel('Accuracy (%)')
axes[0].set_title('Model Accuracy Comparison', fontweight='bold')
axes[0].set_ylim([0, 100])
for i, v in enumerate(model_accs):
    axes[0].text(i, v + 1, f'{v:.2f}%', ha='center', fontweight='bold')

# Plot 2: Confusion Matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Denied', 'Approved'],
            yticklabels=['Denied', 'Approved'], ax=axes[1])
axes[1].set_title(f'{best_model_name} - Confusion Matrix', fontweight='bold')
axes[1].set_ylabel('Actual')
axes[1].set_xlabel('Predicted')

plt.tight_layout()
plt.show()

# STEP 11: Example Prediction
print("\n" + "="*60)
print("EXAMPLE PREDICTION")
print("="*60)
sample = X_test.iloc[0:1]
prediction = rf.predict(sample)[0]
proba = rf.predict_proba(sample)[0]
actual = y_test[0]

result = "Approved" if prediction == 1 else "Denied"
actual_result = "Approved" if actual == 1 else "Denied"
confidence = proba[prediction] * 100

print(f"Actual Status: {actual_result}")
print(f"Predicted Status: {result}")
print(f"Confidence: {confidence:.1f}%")
print(f"Result: {'Correct' if prediction == actual else 'Incorrect'}")

print("\n" + "="*60)
print("PROJECT COMPLETE!")
print("="*60)
print(f"\n Final Results:")
print(f"Best Model: {best_model_name}")
print(f"Test Accuracy: {accuracies[best_model_name]*100:.2f}%")
print(f"Total Samples: {len(X_test)} test applications")
print(f"Correct Predictions: {(predictions[best_model_name] == y_test).sum()}")
print(f"Incorrect Predictions: {(predictions[best_model_name] != y_test).sum()}")
```
